```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# calling read_csv to import a csv file
# create a DataFrame
df = pd.read_csv('Malaysian_Tourism_Data_1995_to_2019nn.csv')
df
```

```python
#axis: If we want to aggregate the columns, then we’ll use axis=1. For rows we’ll use axis=0.
#numeric_only = take only numeric columns.

df['total'] = df.sum(axis=1, numeric_only= True)

df
```

```python
df.iloc[df['total'].idxmax(), 0]
```

```python
df['total'].describe()
```

```python
df['total'].plot()
```

```python
fig, ax = plt.subplots(figsize=(100, 20))
x = list(df['Country Code'])  #take country name for x
y = list(df['total'])         #display value total for y
plt.xlabel('Country Code')    #label x axis
plt.ylabel('Total')           #label y axis
ax.plot(x,y, linewidth = 2)  #plot and set linewidth
plt.show()
```
